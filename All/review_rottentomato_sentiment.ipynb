```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.rottentomatoes.com/m/kung_fu_panda_4/reviews'
```

```python
page = requests.get(link)
page
```

```python
page.content
```

```python
soup= bs(page.content)
```

```python
soup.prettify()
```

```python
names = soup.find_all('a',class_='display-name')
names
```

```python
cust_name = []
for i in range(0,len(names)):
    cust_name.append(names[i].get_text().replace("  ", "").replace("\n", ""))
cust_name
```

```python
len(cust_name)
```

```python
text = soup.find_all('p',class_='review-text')
text
```

```python
reviews = []
for i in range(0,len(text)):
    reviews.append(text[i].get_text())
reviews
```

```python
len(reviews)
```

```python
reviews[:] = [titles.lstrip('\n') for titles in reviews]
reviews
```

```python
reviews[:] = [texts.rstrip('\n') for texts in reviews]
reviews
```

```python
rating = soup.find_all('score-icon-critic-deprecated')
rating
```

```python
rate = []
for i in range(0,len(rating)):
    if rating[i].get("state") == "rotten":
      rate.append("Negative")
    else:
      rate.append("Positive")
rate
```

```python
len(rate)
```

```python
import pandas as pd
```

```python
df1 = pd.DataFrame()
df2 = pd.DataFrame()
df3 = pd.DataFrame()
```

```python
df1['Customer Name']=cust_name
df2['Reviews']=reviews
df3['Ratings']=rate
```

```python
result = pd.concat([df1,df2,df3],axis=1)
```
