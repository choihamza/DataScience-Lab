```python
edu = pd.read_csv('educ_figdp_1_Data.csv',
                  na_values=':', usecols=['TIME', 'GEO', 'Value'])
edu
```

```python

# Replace NaN in Value with mean

edu['Value'].fillna(edu["Value"].mean(), inplace=True)

edu
```

```python
# 2 new columns
# scaleval1=(value in that row -value.min())/(value.max()-value.min())
# scaleval2 =(value in that row-value.mean())/value.standard_deviation()


edu['scaleval1'] = (edu['Value'] - edu['Value'].min()) / (edu['Value'].max() - edu['Value'].min())
edu['scaleval2'] = (edu['Value'] - edu['Value'].mean()) / edu['Value'].std()

edu

# save edu to csv
edu.to_csv('Lab1_200789.csv',index=False)

# edu[['scaleval1','scaleval2']].to_csv('test_200789_task.csv', index=False)
```
