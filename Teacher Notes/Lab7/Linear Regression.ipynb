# Linear Regression

In this notebook we will see how regression analysis can help to understand the data behavior, to predict data values (continuous or discrete), and to find important predictors (sparse models). We present four different regression models: Simple linear regression, Multiple linear regression, Polynomial regression and Logistic regression. We evaluate the analysis results qualitatively by means of Seaborn visualization tools and quantitatively by means Scikit-learn library, as well as other toolboxes.

```python
# Computing a and b values without mean
n=6
no1 =(n*np.sum(X * Y)) - (np.sum(X) * np.sum(Y))
di1= (n * np.sum(X **2)) - (np.sum(X)**2)
b=  no1/ di1
no2 =(np.sum(X **2) * np.sum(Y)) - (np.sum(X) * np.sum(X * Y))
di2= (n * np.sum(X **2)) - (np.sum(X)**2)
a=  no2/ di2
print("Slope (b):", b)
print("Intercept (a):", a)
```

```python
# Function to calculate m and b
def linear_regression(X, y):
    x_mean = np.mean(X)
    y_mean = np.mean(y)
    numerator = np.sum((X - x_mean) * (y - y_mean))
    denominator = np.sum((X - x_mean) ** 2)
    b = numerator / denominator
    a = y_mean - (b * x_mean)
    return a, b
```

```python
# Function to calculate prediction
def predict(X, a, b):
    return b * X + a
```

```python
# Function to calculate RMSE
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

```python
X = np.array([7, 8, 10, 12, 15, 18])
Y = np.array([9, 10, 12, 13, 16, 20])
```

```python
# Training the model
a,b = linear_regression(X, Y)
```

```python
# Making predictions
predictions = predict(X, a, b)
```

```python
# Calculating RMSE
error = rmse(Y, predictions)
```

```python
print("Slope (b):", b)
print("y-Intercept (a):", a)
print("Predictions:", predictions)
print("RMSE:", error)
```

```python
import matplotlib.pyplot as plt
 
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, predictions, color='red', label='Regression Line')
plt.xlabel('Independent variable X')
plt.ylabel('Dependent variable y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```
